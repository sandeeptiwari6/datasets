## Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams["figure.figsize"] = "25,10"
plt.rcParams["legend.fontsize"] = 16
plt.rcParams["axes.labelsize"] = 16

import os
import csv
```

## Loading data

```python
w = []
x = []
y = []
z = []
FT1 = []
FT2 = []
FT3 = []
FT4 = []
i = 0

# How many times points are needed, i.e. sampling frequency
samplingFrequency = 20480;
# at what intervals time points are sampled
samplingInterval = 1 / samplingFrequency;
# begin time period of the signals
beginTime = 0;
# end time period of the signals
endTime = 1;
# time points
time = np.arange(beginTime, endTime, samplingInterval);
frequencies = 0;

dir_path = "../data/IMS/2nd_test/"
lst = os.listdir(dir_path)
lst.sort()

Limit = 0
MaxFrame = 2000
```

```python
lst[0]
```

```python
for file in lst:
    print(len(file))
    print(os.path.join(dir_path, file))
    
    if Limit > MaxFrame:
        break
    Limit += 1
    
    if len(file) == 19:
        w.append([])
        x.append([])
        y.append([])
        z.append([])
        
        with open(os.path.join(dir_path, file), 'r') as csvfile:
            plots = csv.reader(csvfile, delimiter='\t')
            for row in plots:
                w[i].append(float(row[0]))
                x[i].append(float(row[1]))
                y[i].append(float(row[2]))
                z[i].append(float(row[3]))
        
        # frequency domain representation
        fourierTransform = np.fft.fft(w[i]) / len(w[i]) # normalize amplitude
        fourierTransform = abs(fourierTransform[range(int(len(w[i]) / 2))]) # excluding sampling frequency
        FT1.append(fourierTransform)
        
        # frequency domain representation
        fourierTransform = np.fft.fft(x[i]) / len(x[i]) # normalize amplitude
        fourierTransform = abs(fourierTransform[range(int(len(x[i]) / 2))]) # excluding sampling frequency
        FT2.append(fourierTransform)
        
        # frequency domain representation
        fourierTransform = np.fft.fft(y[i]) / len(y[i]) # normalize amplitude
        fourierTransform = abs(fourierTransform[range(int(len(y[i]) / 2))]) # excluding sampling frequency
        FT3.append(fourierTransform)
        
        # frequency domain representation
        fourierTransform = np.fft.fft(z[i]) / len(z[i]) # normalize amplitude
        fourierTransform = abs(fourierTransform[range(int(len(z[i]) / 2))]) # excluding sampling frequency
        FT4.append(fourierTransform)
        
        tpCount = len(w[i])
        vals = np.arange(int(tpCount / 2))
        timePeriod = tpCount / samplingFrequency
        frequencies = vals / timePeriod
        
        i += 1
```

```python
x = frequencies
y = range(i)

Z1 = np.asarray(FT1)
Z2 = np.asarray(FT2)
Z3 = np.asarray(FT3)
Z4 = np.asarray(FT4)
```

```python
X, Y = np.meshgrid(x, y)

fig1 = plt.figure(1)
aw = Axes3D(fig1, zlim=(0, 0.06))
aw.set_xlabel("Frequency")
aw.set_ylabel("Sample")
aw.set_zlabel("Amplitude")
surf1 = aw.plot_surface(X, Y, Z1, linewidth=0, antialiased=True, cmap=cm.jet)
fig1.colorbar(surf1, shrink=0.4, aspect=10)
```

```python
fig2 = plt.figure(2)
ax = Axes3D(fig2, zlim=(0, 0.06))
ax.set_xlabel("Frequency")
ax.set_ylabel("Sample")
ax.set_zlabel("Amplitude")
surf2 = ax.plot_surface(X, Y, Z2, linewidth=0, antialiased=True, cmap=cm.jet)
fig2.colorbar(surf2, shrink=0.4, aspect=10)
```

```python
fig3 = plt.figure(3)
ay = Axes3D(fig3, zlim=(0, 0.06))
ay.set_xlabel("Frequency")
ay.set_ylabel("Sample")
ay.set_zlabel("Amplitude")
surf3 = ay.plot_surface(X, Y, Z3, linewidth=0, antialiased=True, cmap=cm.jet)
fig3.colorbar(surf3, shrink=0.4, aspect=10)
```

```python
fig4 = plt.figure(4)
az = Axes3D(fig4, zlim=(0, 0.06))
az.set_xlabel("Frequency")
az.set_ylabel("Sample")
az.set_zlabel("Amplitude")
surf4 = az.plot_surface(X, Y, Z4, linewidth=0, antialiased=True, cmap=cm.jet)
fig4.colorbar(surf4, shrink=0.4, aspect=10)
```

```python
data_dir = '../data/IMS/2nd_test'
test_2nd = pd.DataFrame()

for filename in os.listdir(data_dir):
    dataset=pd.read_csv(os.path.join(data_dir, filename), sep='\t')
    dataset_mean_abs = np.array(dataset.abs().mean())
    dataset_mean_abs = pd.DataFrame(dataset_mean_abs.reshape(1,4))
    dataset_mean_abs.index = [filename]
    test_2nd = test_2nd.append(dataset_mean_abs)

test_2nd.columns = ['Bearing 1','Bearing 2','Bearing 3','Bearing 4']
```
